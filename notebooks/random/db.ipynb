```python
import os, sys
cwd = os.getcwd()
project_path = cwd[:cwd.find('pygents')+7]
if project_path not in sys.path: sys.path.append(project_path)
os.chdir(project_path)

import uuid

if 'pygents.db' in sys.modules:
    del sys.modules['pygents.db']

import pygents.db as db
```

```python
# Generate a random UUID once (e.g., via uuid.uuid4()) and use that value.
PSYMINDER_NAMESPACE = uuid.UUID("7a67bc69-ca5e-40c4-ba9b-10b6bf1d5c28") # Example UUID

def create_deterministic_username_id(username):
    """
    Generates a consistent, unique UUID (v5) for a given username.
    """
    # Use a consistent namespace (e.g., uuid.NAMESPACE_DNS)
    # The output will always be the same for "johndoe123"
    unique_id = uuid.uuid5(PSYMINDER_NAMESPACE, username)
    return str(unique_id)

# Example Usage:
user_id_1 = create_deterministic_username_id("johndoe123")
user_id_2 = create_deterministic_username_id("johndoe123")
print(f"User ID 1: {user_id_1}")
print(f"User ID 2: {user_id_2}") # This will be the same as User ID 1
```

```python
pmdb = db.BotMinderDB('psyminder.db',PSYMINDER_NAMESPACE)
pmdb.create()
pmdb.add_user(123456)
pmdb.add_user(123457)
pmdb.add_user(123458)
print(pmdb.get_user(123457))
print(pmdb.set_user(123457))
pmdb.set_user(123457,role = "admin")
pmdb.set_user(123457,metrics_onoff = "on", metrics_threshold = 0.5)
print(pmdb.get_user(123457))
```
