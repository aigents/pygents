```python
import os, sys
cwd = os.getcwd()
project_path = cwd[:cwd.find('pygents')+7]
if project_path not in sys.path: sys.path.append(project_path)
os.chdir(project_path)

from PIL import Image
import fitz #https://pymupdf.readthedocs.io/en/

if 'pygents.text_matching' in sys.modules:
    del sys.modules['pygents.text_matching']

from pygents.text_matching import FuzzyMatcher

```

```python
def convert_pdf_doc_page_to_image(doc, page_num, output_image_path):
    page = doc[page_num]  # Get the specified page (0-indexed)
    # https://pymupdf.readthedocs.io/en/latest/page.html#Page.get_pixmap
    pix = page.get_pixmap(dpi=200)  # Render page to a pixmap (image)
    pix.save(output_image_path)  # Save the image

def convert_pdf_path_page_to_image(pdf_path, page_num, output_image_path):
    doc = fitz.open(pdf_path)
    convert_pdf_page_to_image(doc, page_num, output_image_path)
    doc.close()

def convert_pdf_page_to_image(pdf_path, page_num, output_image_path):
    doc = fitz.open(pdf_path)
    page = doc[page_num]  # Get the specified page (0-indexed)
    # https://pymupdf.readthedocs.io/en/latest/page.html#Page.get_pixmap
    pix = page.get_pixmap(dpi=200)  # Render page to a pixmap (image)
    pix.save(output_image_path)  # Save the image
    doc.close()
```

```python
doc = fitz.open(path)
page = doc[0]  # Access the first page (0-based index)
#https://pymupdf.readthedocs.io/en/latest/document.html#Document.get_toc
toc = doc.get_toc(simple=True)
toc_texts = [i[1] for i in toc]
pagenos = {}
for i in toc:
    pagenos[i[1]] = i[2]
print(toc_texts)
print(pagenos)
doc.close()
```

```python
fm = FuzzyMatcher(toc_texts)
print(fm.match('кбт'))
print(fm.match('тревожные расстройства'))
print(fm.match('когнитивные ошибки'))
print(fm.match('когнитивные искажения'))

print(fm.match('тревожные расстройства')[0])
print(pagenos[fm.match('тревожные расстройства')[0]])

print(fm.match('оперантное обуславливание')[0])
print(pagenos[fm.match('оперантное обуславливание')[0]])

convert_pdf_page_to_image(path,pagenos[fm.match('оперантное обуславливание')[0]]-1,"temp.png")

# Load an image
img = Image.open("temp.png")

# Display the image (opens in default image viewer)
img.show()
```
